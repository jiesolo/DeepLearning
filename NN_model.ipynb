```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.seqential = nn.Sequential(nn.Conv2d(3, 32, 5, padding=2),
                                      nn.MaxPool2d(2),
                                      nn.Conv2d(32,32,5,padding=2),
                                      nn.MaxPool2d(2),
                                      nn.Conv2d(32,64,5,padding=2),
                                      nn.MaxPool2d(2),
                                      nn.Flatten(),
                                      nn.Linear(1024,64),
                                      nn.ReLU(),
                                      nn.Linear(64,10),
                                      nn.Sigmoid()
                                      )
    def forward(self, x):
        x = self.seqential(x)
        return x
```
